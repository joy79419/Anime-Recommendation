```python
import pandas as pd
data = pd.read_csv('animelists_reduced_svd.csv')
data[:5]
```

```python
class_df = pd.read_csv('classmate_data.csv')
```

```python
class_df['username'] = class_df['user_id']
class_df['my_score'] = class_df['rating']
class_df = class_df.iloc[:,2:]
```

```python
final_df = pd.concat([data,class_df], axis=0)
```

```python
df = final_df[['username','anime_id','my_score']]
df['user_id'] = pd.factorize(df['username'])[0]
```

```python
df=df.iloc[:,[3,0,1,2]]
```

```python
df = df.reset_index().iloc[:,[1,2,3,4]]
```
